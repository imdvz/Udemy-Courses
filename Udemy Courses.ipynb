```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading CSV FIle.

```python
df = pd.read_csv('G:/DOCUMENTS/udemy_courses.csv')
```

### Filling zeros in NaNs and making a new column.

```python
df.fillna(0)

# Making a new Column
df['Count'] = 1
df.head(3)
```

## Which category has the biggest 'content_duration'?

### Grouping by Level.

```python
level = df.groupby('level').sum()
level
```

```python
# I HAVE SHOWN THE BAR OF "All Levels" SO THAT BY ADDING OTHER BARS ONE CAN ROUGHLY VERIFY THAT EVERYTHING IS RIGHT.

# Making a list of elements present in column "level" and storing them in "level1".
level1 = [level for level, x in level.groupby('level')]

plt.figure(figsize=(10,5))
plt.bar(level1, level['content_duration'])
plt.xlabel('Category', size=13)
plt.ylabel('Content Duration(in min)', size=13)
plt.yticks(np.array([0 ,500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000]))

plt.show()
```

## Which category has the most number of lectures?

```python
# I HAVE SHOWN THE BAR OF "All Levels" SO THAT BY ADDING OTHER BARS ONE CAN ROUGHLY VERIFY THAT EVERYTHING IS RIGHT.

## plt.figure(figsize=(10,5))
plt.bar(level1, level['num_lectures'])
plt.xlabel('Category', size=13)
plt.ylabel('Number of Lectures', size=13)
plt.yticks(np.array([0, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000]))

plt.show()
```

## Which Category has the most number of Subscribers?

```python
# I HAVE SHOWN THE BAR OF "All Levels" SO THAT BY ADDING OTHER BARS ONE CAN ROUGHLY VERIFY THAT EVERYTHING IS RIGHT.

plt.figure(figsize=(10,5))
plt.bar(level1, level['num_subscribers'])
plt.yticks(np.array([0 , 500000, 1000000, 2000000, 3000000, 4000000, 5000000, 6000000, 7000000]))

plt.show()
```

## How many Courses are available in each category?

```python
# I HAVE SHOWN THE BAR OF "All Levels" SO THAT BY ADDING OTHER BARS ONE CAN ROUGHLY VERIFY THAT EVERYTHING IS RIGHT.

plt.figure(figsize=(8,4))
plt.bar(level1, level['Count'])
plt.ylabel('Number of Courses', size=13)
plt.xlabel('Category', size=13)
plt.show()
```

## How many courses are Paid and how many are not?

### Grouping by is_paid.

```python
isPaid = df.groupby('is_paid').sum()
isPaid
```

```python
plt.bar(['False', 'True'], isPaid['Count'])
plt.xlabel('Paid?',size=13)
plt.ylabel('Count', size=13)

plt.show()
```

## How many courses are available for Different Subjects?

### Grouping by Subject.

```python
subject = df.groupby('subject').sum()
subject
```

```python
subject1 = [subject for subject, x in subject.groupby('subject')]

plt.figure(figsize=(8,4))
plt.bar(subject1, subject['Count'])
plt.xlabel('Subject', size=13)
plt.ylabel('Count', size=13)
plt.yticks(np.arange(0,1201,100))
plt.show()
```
